```python
import pandas as pd

# Define the possible values for A and B (True and False)
A = [False, False, True, True]
B = [False, True, False, True]

# Create a dictionary to store the data
data = {
    'A': A,
    'B': B,
    'A&B': [a and b for a, b in zip(A, B)],
    'A|B': [a or b for a, b in zip(A, B)],
    'A^B': [a ^ b for a, b in zip(A, B)]
}

# Create a pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Display the DataFrame
display(df)
```
